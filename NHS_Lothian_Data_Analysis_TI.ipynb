# Import packages and dataclean function

```python
from google.colab import drive
from pathlib import Path
drive.mount('/content/drive')
lothian = 'S08000024'
```

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
from pathlib import Path

def dataclean(file):
  file = file.loc[file['HBT'] == lothian]
  # Keeps data from 20190101 (inclusive) to 20250101 (exclusive)
  if 'QuarterEnding' in file.columns:
    file = file.loc[(file['QuarterEnding'] >= 20190101) & (file['QuarterEnding'] < 20260101)]
    file['QuarterEnding'] = pd.to_datetime(file['QuarterEnding'].astype(str), format='%Y%m%d')
  elif 'MonthEnding' in file.columns:
    file = file.loc[(file['MonthEnding'] >= 20190101) & (file['MonthEnding'] < 20260101)]
    file['MonthEnding'] = pd.to_datetime(file['MonthEnding'].astype(str), format='%Y%m%d')
  elif 'MonthEnd' in file.columns:
    file = file.loc[(file['MonthEnd'] >= 20190101) & (file['MonthEnd'] < 20260101)]
    file['MonthEnd'] = pd.to_datetime(file['MonthEnd'].astype(str), format='%Y%m%d')
  return file
```

```python
selected_wait_columns = ['LessThan4WeekWait', '4To8WeekWait', '8To12WeekWait']
analysis1a = sot_distribution_of_completed_waits[selected_wait_columns].apply(pd.to_numeric, errors='coerce').fillna(0)
analysis1a = analysis1a.groupby(sot_distribution_of_completed_waits['QuarterEnding']).sum()

analysis1a['0To12WeekWait'] = analysis1a['LessThan4WeekWait'] + analysis1a['4To8WeekWait'] + analysis1a['8To12WeekWait']

plt.figure(figsize=(12, 6))
plt.plot(analysis1a.index, analysis1a['0To12WeekWait'])
plt.xlabel('Quarter Ending')
plt.ylabel('Total Completed Waits (0-12 Weeks)')
plt.title('Completed Waits (0-12 Weeks) by Quarter')
plt.legend(['0-12 Weeks Total'], bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Heatmap

```python
specialty_codes = pd.read_csv('/content/drive/MyDrive/PDA Data Science SCQF Level 8/Project/Raw Data/specialty_codes.csv')

specialty_codes['CompletedCounts'] = 0
specialty_codes['OngoingCounts'] = 0

for i in range(len(specialty_codes)):
  find_code = specialty_codes['Specialty'][i]
  specialty_codes.loc[i, 'CompletedCounts'] = specialty_counts_completed_series.get(find_code, 0)
  specialty_codes.loc[i, 'OngoingCounts'] = specialty_counts_ongoing_series.get(find_code, 0)

comparison = specialty_codes.reset_index(drop=True)
comparison = comparison.drop('Specialty', axis=1)
```

```python
correlation = comparison[['CompletedCounts', 'OngoingCounts']].corr()
```

```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.heatmap(correlation, cmap='Blues', vmin=0.9, vmax=1, annot=True, fmt='.5f')
plt.title('Correlation between Completed and Ongoing Waits')
plt.show()
```

# Ongoing Waits Modelling

```python
x = analysis2a.index.astype(np.int64) // 10**9
y = analysis2a['0To12WeekWait']
a, b = np.polyfit(x, y, 1)
plt.scatter(x, y)
plt.plot(x, a*x+b)
print(f"a = {a}, b = {b}")
```

```python
predict = pd.Timestamp('2026-12-31').to_datetime64().astype(np.int64)
prediction_2a = a * predict + b
print(f"Predicted 0To12WeekWait on 2026-12-31: {prediction_2a:.0f}")
```

```python
further_predict = pd.Timestamp('2027-12-31').to_datetime64().astype(np.int64)
further_prediction_2a = a * further_predict + b
print(f"Predicted 0To12WeekWait on 2027-12-31: {further_prediction_2a:.0f}")
```
